<h1><b> Grayscale conversion <b></h1>

```python
# Grayscale conversion
gray = cv.cvtColor(img_mtlb,cv.COLOR_BGR2GRAY)
cv.imshow('image', gray)
cv.waitKey(0)
```

```python
# RGB[A] to Gray:Y0.299*R+0.587*G+0.114*B
# coefficients = [0.114, 0.587, 0.299] [R,g,b]
```

```python
#Using formula
coefficients = [0.114, 0.587, 0.299]
m = np.array(coefficients).reshape((1,3))
gray_cvf = cv.transform(img,m)
cv.imshow('image', gray_cvf)
cv.waitKey(0)
#cv.destroyAllWindows()
```

<h1><b> Resizing Image with Opencv <b></h1>

```python
#Resizing image using opencv
W, H, D = img.shape  # original dimension of the image
print(W, H, D)
width, height= (320,320)
img_resize = cv.resize(img,(width,height),interpolation = cv.INTER_LINEAR)
normalized = cv.normalize(img_resize, img_resize, 0, 320, cv.NORM_MINMAX)
cv.imshow('image', normalized)
cv.waitKey(0)
```

```python
img_resize = cv.resize(img, None, fx=2,fy=2,interpolation=cv.INTER_LINEAR) #Transform dimension
# cv.INTER_LINEAR for zooming image
cv.imshow('image',img_resize)
cv.waitKey(0)
cv.destroyAllWindows()
```

<h1><b> Rotation of an Image using opencv<b></h1>

```python
#Grab the dimension of the image and calculate the center of the image
(h,w) = img.shape[:2]
center = (w / 2, h / 2)
#print(center)

#Rotate the image by 180 degrees
M = cv.getRotationMatrix2D(center, 18,1.0)
rotated = cv.warpAffine(img, M, (w,h))
cv.imshow('rotated',rotated)
cv.waitKey(0)
```

<h1><b> Cropping an Image using opencv<b></h1>

```python
# Cropping the image using aray slices -- it's a Numpy array
cropped = img[20:180, 30:150]
cv.imshow("cropped",cropped)
cv.waitKey(0)
```
